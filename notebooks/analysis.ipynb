```python
def detect_outliers_iqr(data, feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[feature] < lower) | (data[feature] > upper)]

outliers = {}
for battery_id in battery_list:
    for f in features:
        outliers[f] = detect_outliers_iqr(df[df['battery_id']==battery_id], f)
        print(f"for {battery_id} {f}: {len(outliers[f])} outliers")
```
